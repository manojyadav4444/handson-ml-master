```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC
```

```python
iris = datasets.load_iris()
```

```python
X = iris['data'][:,(2,3)]
y = ( iris['target'] == 2 ).astype(np.float64)
```

```python
svm_clf = Pipeline((
    ('scaler', StandardScaler()),
    ('lin_svc', LinearSVC(C=1, loss='hinge'))
))
svm_clf.fit(X,y)
svm_clf.predict([[5.5, 1.7]])
```

```python
from sklearn.datasets import make_moons
from sklearn.preprocessing import PolynomialFeatures
make_moons
X, y = datasets.make_moons(n_samples = 300, shuffle = True, noise = 0.05, random_state = 42)
X, y
```

```python
from sklearn.svm import LinearSVC

poly_svm_clf = Pipeline((
    ('poly_feat', PolynomialFeatures(degree = 3)),
    ('scaler', StandardScaler()),
    ('svm_clf', LinearSVC(C = 10, loss = 'hinge'))
))
poly_svm_clf.fit(X,y)
poly_svm_clf.predict([[-5.06952598e-01,  8.43251391e-01]])
```

```python
from sklearn.svm import SVC

poly_kernel_svm_clf = Pipeline((
    ('scaler', StandardScaler()),
    ('svc_clf', SVC(kernel='poly', degree = 3, coef0 = 1, C = 5 ))
))
poly_kernel_svm_clf.fit(X, y)
poly_kernel_svm_clf.predict([[6.22519125e-01, -3.72101344e-01]])
```

```python
from sklearn.svm import SVC

rbf_kernel_svm_clf = Pipeline((
    ('scaler', StandardScaler()),
    ('svm_clf', SVC(kernel = 'rbf', gamma = 5, C = 0.001))
))
rbf_kernel_svm_clf.fit(X, y)
rbf_kernel_svm_clf.predict([[6.22519125e-01, -3.72101344e-01]])
```

```python
np.random.seed(42)
X = np.random.randn(100,1)
y = 4 + 3*X + np.random.randn(100,1)
print(f'{X}\n\n\n\n')
print(y)
```

```python
from sklearn.svm import LinearSVR

lin_svm_reg = Pipeline((
    ('scaler', StandardScaler()),
    ('lin_reg', LinearSVR(epsilon = 1.5))
))
```

```python
lin_svm_reg.fit(X,y)
lin_svm_reg.predict([[0.49671415]])
```

```python
from sklearn.svm import SVR

poly_svm_reg = Pipeline((
    ('scaler', StandardScaler()),
    ('svm_reg', SVR(kernel = 'poly', degree= 3, C=100, epsilon = 0.1))
))
poly_svm_reg.fit(X,y)
poly_svm_reg.predict([[0.49671415]])
```
